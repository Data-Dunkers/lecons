## Abandon d'une ligne par numéro d'index

Vous remarquez qu'il reste cette dernière ligne, qui contient "TOT" pour l'abréviation de l'équipe ? Supprimons cette ligne (ligne 9).

```python
display(df.drop(9))
```

Que se passe-t-il si nous remplaçons le "9" par un autre chiffre ? Essayez-le !

Affichons à nouveau le cadre de données pour voir si le changement est permanent.

```python
display(df)
```

Ce n'est pas le cas. La ligne "Carrières" est toujours là.

Nous pouvons la rendre permanente en la réaffectant à la variable DataFrame.

```python
df = df.drop(9)

display(df)
```

## Tri

Que se passe-t-il si nous voulons trier les données par vol (STL) ?

```python
display(df.sort_values('STL'))
```

Et si nous voulions qu'ils soient décroissants au lieu d'être croissants ? Ajoutez simplement `ascending=False` aux arguments de `sort_values()`. (La valeur par défaut, si vous ne lui dites pas autre chose, est `ascending=True`)

```python
display(df.sort_values('STL', ascending=False))
```

Réduisons les colonnes que nous regardons et sauvegardons-les dans un nouveau dataframe nommé `df2`.

Si nous devions continuer à travailler avec ces seules colonnes, nous travaillerions alors avec `df2` à partir de maintenant.

```python
columns = ['SEASON_ID', 'TEAM_ABBREVIATION', 'GP', 'GS', 'BLK', 'STL', 'MIN', 'FGM', 'FGA']
df_2 = df[columns]
display(df_2)
```

Nous travaillerons désormais avec `df2`.

Trions maintenant sur deux colonnes, par exemple d'abord sur les Blocs par match (BLK) et ensuite sur les Volées par match (STL). Notez que nous devons mettre les noms des colonnes dans une liste (`[ ]`).

```python
display(df_2.sort_values(['BLK', 'STL']))
```

Pouvez-vous confirmer que le tri s'est effectué correctement ?
